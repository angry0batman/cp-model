```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("cp_scores_adjusted.csv")
df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
X = df[['CODECHEF','CODEFORCES','LEETCODE']]
y = df['CP_SCORE']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=100)
```

```python
model=LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred= model.predict(X_test)
```

```python
print(y_pred)
```

```python
#mean_squared_error, r2_score
print("mean squared error value is:",mean_squared_error(y_test,y_pred))
```

```python
print("r2 score is:",r2_score(y_test, y_pred))
```

```python
import joblib
joblib.dump(model, 'CP.joblib')
```

```python
loaded_model = joblib.load('CP.joblib')
```

```python
new_data = np.array([[1400,2000,1300]])


predictions = loaded_model.predict(new_data)

# Print predictions
print("Predictions:")
print(f"Predicted value: {predictions[0]}")
```
